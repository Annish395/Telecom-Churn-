```python
# Removing Non Essential columns from categorical Columns
# 1. Service Area, 2. OptOutMailings, 3. NonUSTravel, 4. NotNewCellphoneUser, 5. HandsetPrice
```

```python
from sklearn.preprocessing import StandardScaler
import statsmodels.api as sm
```

```python
df1 = df.copy()
```

```python
df1.drop(['ServiceArea','OptOutMailings','NonUSTravel','NotNewCellphoneUser','HandsetPrice'],axis=1,inplace=True)
```

```python
num_df
```

```python
# Removing non-essential columns from Numerical Columns
# MonthlyMinutes, PercChangeMinutes, HandsetModels

df1.drop(['MonthlyMinutes','PercChangeMinutes','HandsetModels'],axis=1,inplace=True)
```

```python
df1.shape
```

```python
df1.drop('OwnsMotorcycle',axis=1,inplace=True)
```

```python
df1.shape
```

```python
df1['RoamingCalls'] = df1['RoamingCalls'].apply(lambda x : 0 if x == 0.0 else 1)
df1['ThreewayCalls'] = df1['ThreewayCalls'].apply(lambda x : 0 if x == 0.0 else 1)
df1['OverageMinutes'] = df1['OverageMinutes'].apply(lambda x : 0 if x == 0.0 else 1)
df1['DirectorAssistedCalls'] = df1['DirectorAssistedCalls'].apply(lambda x : 0 if x == 0.0 else 1)
df1['CustomerCareCalls'] = df1['CustomerCareCalls'].apply(lambda x : 0 if x == 0.0 else 1)
df1['CallForwardingCalls'] = df1['CallForwardingCalls'].apply(lambda x : 0 if x == 0.0 else 1)
```

```python
df1
```

```python
#1. Yes  & No --> Churn, ChildrenInHH, HandsetRefurbished, HandsetWebCapable, TruckOwner, RVOwner, BuysViaMailOrder,
#                RespondsToMailOffers, OwnsComputer, HasCredtiCard, NewCellphoneUser, MadeCallToRetentionTeam
encode = {'Yes':1 , 'No' : 0}
encode_1 = {'Known':1 , 'Unknown' : 0}
df1['Churn'] = df1['Churn'].map(encode)
df1['ChildrenInHH'] = df1['ChildrenInHH'].map(encode)
df1['HandsetRefurbished'] = df1['HandsetRefurbished'].map(encode)
df1['HandsetWebCapable'] = df1['HandsetWebCapable'].map(encode)
df1['TruckOwner'] = df1['TruckOwner'].map(encode)
df1['RVOwner'] = df1['RVOwner'].map(encode)
df1['Homeownership'] = df1['Homeownership'].map(encode_1)
df1['BuysViaMailOrder'] = df1['BuysViaMailOrder'].map(encode)
df1['RespondsToMailOffers'] = df1['RespondsToMailOffers'].map(encode)
#df2['OptOutMailings'] = df2['OptOutMailings'].map(encode)
#df2['NonUSTravel'] = df2['NonUSTravel'].map(encode)
df1['OwnsComputer'] = df1['OwnsComputer'].map(encode)
df1['HasCreditCard'] = df1['HasCreditCard'].map(encode)
df1['NewCellphoneUser'] = df1['NewCellphoneUser'].map(encode)
#df2['NotNewCellphoneUser'] = df2['NotNewCellphoneUser'].map(encode)
#df2['OwnsMotorcycle'] = df2['OwnsMotorcycle'].map(encode)
df1['MadeCallToRetentionTeam'] = df1['MadeCallToRetentionTeam'].map(encode)
```

```python
df1
```

```python
# DummyEncoding --> IncomeGroup, CreditRating, PrizmCode, Occupation, MaritalStatus
df2 = pd.get_dummies(df1[['IncomeGroup','CreditRating','Occupation','MaritalStatus']],drop_first=True)
```

```python
df1.drop(['IncomeGroup','CreditRating','PrizmCode','Occupation','MaritalStatus'],axis=1,inplace=True)
```

```python
model_df = pd.concat([df1,df2],axis=1)
```

```python
ss = StandardScaler()

model_df['MonthlyRevenue'] = ss.fit_transform(model_df[['MonthlyRevenue']])
model_df['TotalRecurringCharge'] = ss.fit_transform(model_df[['TotalRecurringCharge']])
model_df['PercChangeRevenues'] = ss.fit_transform(model_df[['PercChangeRevenues']])
model_df['DroppedCalls'] = ss.fit_transform(model_df[['DroppedCalls']])
model_df['BlockedCalls'] = ss.fit_transform(model_df[['BlockedCalls']])
model_df['UnansweredCalls'] = ss.fit_transform(model_df[['UnansweredCalls']])
model_df['ReceivedCalls'] = ss.fit_transform(model_df[['ReceivedCalls']])
model_df['OutboundCalls'] = ss.fit_transform(model_df[['OutboundCalls']])
model_df['InboundCalls'] = ss.fit_transform(model_df[['InboundCalls']])
model_df['PeakCallsInOut'] = ss.fit_transform(model_df[['PeakCallsInOut']])
model_df['OffPeakCallsInOut'] = ss.fit_transform(model_df[['OffPeakCallsInOut']])
model_df['DroppedBlockedCalls'] = ss.fit_transform(model_df[['DroppedBlockedCalls']])
model_df['CallWaitingCalls'] = ss.fit_transform(model_df[['CallWaitingCalls']])
model_df['MonthsInService'] = ss.fit_transform(model_df[['MonthsInService']])
model_df['UniqueSubs'] = ss.fit_transform(model_df[['UniqueSubs']])
model_df['ActiveSubs'] = ss.fit_transform(model_df[['ActiveSubs']])
#model_df['Handsets'] = ss.fit_transform(model_df[['Handsets']])
model_df['CurrentEquipmentDays'] = ss.fit_transform(model_df[['CurrentEquipmentDays']])
model_df['AgeHH1'] = ss.fit_transform(model_df[['AgeHH1']])
model_df['AgeHH2'] = ss.fit_transform(model_df[['AgeHH2']])
```

```python
X = model_df.drop('Churn',axis=1)
X = sm.add_constant(X)
y = model_df['Churn']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=123)
```

```python
log_model = sm.Logit(y_train,X_train).fit()
print(log_model.summary())
```
