### 2. MonthlyRevenue

```python
total_monthlyrevenue = df.MonthlyRevenue.sum()

revenue_fromChurners = df[df.Churn == 'Yes']['MonthlyRevenue'].sum()
print('Total Monthly Revenue of the Company : ',(df['MonthlyRevenue'].sum()))
print('Potential Loss if those who said Yes to Churn Leave : ',revenue_fromChurners )

percentage_loss = round(revenue_fromChurners * 100 / total_monthlyrevenue,3)
print('Percentage loss in Monthly revenue if Churners leave : ',percentage_loss,'%')
```

```python
an_df = pd.DataFrame([{'Total': total_monthlyrevenue,'Potential Loss from Churners' : revenue_fromChurners,
                      'Potential Loss Percentage' : percentage_loss}],index=['Monthly Revenue'])
an_df
```

### 3. MonthlyMinutes

```python
total_monthlymins = df.MonthlyMinutes.sum()
print('Total Minutes Spent on Call per Month : ',total_monthlymins)

minutes_churners = df[df['Churn'] == 'Yes']['MonthlyMinutes'].sum()
print('Potential Loss in Minutes if those who said Yes to Churn Leave : ',minutes_churners)

percent_minutesloss = round(minutes_churners * 100 / total_monthlymins,2)
print('Potential Loss Percentage : ',percent_minutesloss,'%')
```

```python
minutes_df= pd.DataFrame([{'Total': total_monthlymins,'Potential Loss from Churners' : minutes_churners,
                      'Potential Loss Percentage' : percent_minutesloss}],index=['Monthly Minutes'])
an_df = an_df.append(minutes_df)
an_df
```

# Outliers Treatment

```python
num_df3 = df3.drop('Churn',axis=1)

churn_df3 = df3[['Churn']]
num_df3.shape
```

```python
encode = {'Yes':1 , 'No' : 0}

churn_df3['Churn'] = churn_df3['Churn'].map(encode)
```

```python
churn_df3.isnull().sum()
```

```python
model_df = pd.concat([num_df3,churn_df3],axis=1)
model_df.isnull().sum()
```

# Scaling

```python
from sklearn.preprocessing import StandardScaler
ss = StandardScaler()

X = model_df.drop('Churn',axis=1)
y = model_df[['Churn']]

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=100,test_size=0.3)
```

```python
sc_X_train = pd.DataFrame(ss.fit_transform(X_train),columns=X_train.columns)
sc_X_test = pd.DataFrame(ss.fit_transform(X_test),columns=X_test.columns)
```

# Model Building

```python
from sklearn.linear_model import LogisticRegression
lg = LogisticRegression()

log_model = lg.fit(sc_X_train,y_train)

from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
```

```python
y_pred_train = log_model.predict(sc_X_train)
print('Classification Report for Train Data : \n\n',classification_report(y_train,y_pred_train))
```

```python
y_pred_test = log_model.predict(sc_X_test)
print('Classification Report for Test Data : \n\n',classification_report(y_test,y_pred_test))
```
